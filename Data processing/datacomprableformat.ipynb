```python
import pandas as pd
dataFile = "/content/drive/MyDrive/avalonWHistoryfixed.tsv"
data = pd.read_csv(dataFile, sep = '\t')
```

```python
#comboprompt no his
data['nohiscombo'] = ""

for i, (index, row) in enumerate(data.iterrows()):
    if pd.isna(row['history']):
      x = ''
    else:
      x = row['history']

    prmpt = {
        "name": "combopromptnoHis",

"prompt": """{"Task" : "You will be given a message from the hidden role game Avalon. Your task is to determine whether this message is True or Deceptive. Your prediction must be based on the given example.", \n  "Example of a deceptive message" : "i'm good, so i'll vote yes on this party. i'm not sure about player-3, but i'm not sure about anyone at this point.",\n "Message to predict" : """ + row['withName'] + """, \n "Output Format" : "True" or "Deceptive" }"""
}

    data.at[index, 'nohiscombo'] = prmpt['prompt'] +  ""
print(data['nohiscombo'][3])


```

```python
data['hiscombo'] = ""

for i, (index, row) in enumerate(data.iterrows()):
    if pd.isna(row['history']):
      x = ''
    else:
      x = row['history']

    prmpt = {
        "name": "comboprompt",

"prompt": """{"Task" : "You will be given a message from the hidden role game Avalon. Your task is to determine whether this message is True or Deceptive. Your prediction must be based on the game history and given example.", \n  "Example of a deceptive message" : "i'm good, so i'll vote yes on this party. i'm not sure about player-3, but i'm not sure about anyone at this point." , \n "Game History" : """ + x + """,\n "Message to predict" : """ + row['withName'] + """, \n "Output Format" : "True" or "Deceptive" }"""
}

    data.at[index, 'hiscombo'] = prmpt['prompt'] +  ""
print(data['hiscombo'][0])
```

```python
data.to_csv("/content/drive/MyDrive/combopreppedData.tsv", sep = '\t')
```
